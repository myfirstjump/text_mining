```python
# name    len去掉第一筆: 46
# 第一筆 :'<span class="sr-hotel__name">\\n\', \'Aoi Kyoto Stay（葵京都住宿別墅）
# 其他筆 :customGoal:OLAQMJbQPLbOMFcPRHe:2">\\n\', \'Iori Machiya Stay（庵町家之居度假屋）\\n\'
name = re.findall("customGoal:OLAQMJbQPLbOMFcPRHe:2.>\\\\n\\', \'(.+?)\\\\n\\', \\'<", content)
```

```python
# price   有sellout的，要設空值   keyword: <h3 class="fe_banner__title"> 剛剛售完！
price = re.findall("TWD\\\\xa0(\w+,\w+)\\\\n\\',\s\\'</b>", content)
```

```python
# Latitude and Longitude    len: 完整 47
# map_address_pin" data-coords="135.764805749059,35.0006704657926"
latitude = re.findall('map_address_pin" data-coords="([0-9]+\.[0-9]+),\w[0-9]+\.[0-9]+', content)
longitude = re.findall('map_address_pin" data-coords="[0-9]+\.[0-9]+,(\w[0-9]+\.[0-9]+)', content)
```

```python
# 幾晚    # 同price，空值要處理
# aria-label="4 晚房價
numNight = re.findall('aria-label="(\w 晚房價)', content)
```

```python
# 區域     len: 完整 47
# 顯示地圖">\\n\', \' 上京區, 京都【西陣】\\n\
# 顯示地圖">\\n\', \' 嵐山・高雄, 京都
Region = re.findall('''顯示地圖.>\\\\n.,\s.\s([\w|・]+,\s[\w+|【|】|・|·]+)''', content)
```
