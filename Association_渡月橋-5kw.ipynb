```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
import time
from mlxtend.frequent_patterns import association_rules
import json
import numpy as np
```

```python
src = "渡月橋_Titlize" # DB 讀取來源 collection
filePath = "C:\\Users\\Java\\Desktop\\TM"
focus = "allTags" # 欲加強的目標
multi = 100  # 加強目標的權重
nowtime = "0915"#time.strftime("%m%d")
kwNum = 5
# "渡月橋" Togetsukyō
# "平等院鳳凰堂"  Byōdōin
# "千本鳥居" Fushimi-Inari Taisha
# "京都御所" Kyōto gosho
# "京都塔" kyoto tower
# "二條城" Nijō-jō
# "三十三間堂" Sanjūsangen-dō
# "清水寺" Kiyomizu-dera
# "金閣寺" Rokuonji
```

```python
# load 關鍵字     ->    list atcKwList
f = open(filePath + "\\clustering\\keywordsSets\\{}kw-{}-weight-{}-{}keywords{}.txt".format(kwNum, multi, focus, str(src), nowtime), 'r', encoding="utf-8")
atcKwList = json.load(f)
f.close()
```

```python
# load 關鍵字set   ->   list kws
kws = []
f = open(filePath + "\\clustering\\keywordsSets\\{}kw-{}-weight-{}-{}kwSet{}.txt".format(kwNum, multi, focus, str(src), nowtime), 'r', encoding="utf-8")
kw = f.readlines()
for i in kw:
    kws.append(i.strip())
kws.remove('url')
kws.remove('title')
f.close()
```

```python
# np.zeros( (3,4) )
table = np.zeros( (len(atcKwList), len(kws) ) )
table.shape
```

```python
# 將 table 填入值
for i in range(0, len(atcKwList)):
    for word in atcKwList[i]:
        if word in kws:
            idx = kws.index(word)
            table[i, idx] = 1 # atcKwList[i][word]
```

```python
df = pd.DataFrame(table, columns = kws)
```

```python
frequent_itemsets = apriori(df, min_support=0.005, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.1)
rules
```

```python
siteRule = rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.5)]
```
